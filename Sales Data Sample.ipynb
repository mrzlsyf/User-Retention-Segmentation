### **Data Outliers**
*Outliers data is data or observations that deviate extremely from the average of a set of existing data*

```python
numeric = dataset[['QUANTITYORDERED', 'PRICEEACH', 'SALES']]

def count_outliers(data):
    columns = numeric
    outlier_counts = {}

    for column in columns:
        Q1 = dataset[column].quantile(0.25)
        Q3 = dataset[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        outliers = dataset[(dataset[column] < lower_bound) | (dataset[column] > upper_bound)]
        outlier_counts[column] = len(outliers)

        print(f"Kolom: {column}")
        print(f"  Lower Bound: {lower_bound}")
        print(f"  Upper Bound: {upper_bound}", "\n")

    return outlier_counts

outlier_counts = count_outliers(data)
print(f'Berikut Jumlah Data Outliers yang Ada:', outlier_counts)
```

# **Data Binning and Case Statement**
*Data Binning is the process of combining numeric data into certain categories based on a specified range*

*3 binnings are performed, namely in the QUANTITYORDERED, PRICEEACH, and SALES columns*

### **QUANTITYORDERED**
*The number of orders is divided into 3 bins, namely Small, Medium, and Large. So the results are as follows:*

```python
newdata = dataset.copy()
bins = np.linspace(min(dataset['QUANTITYORDERED']), max(dataset['QUANTITYORDERED']), 4)
category = ['Small', 'Medium', 'Large']

newdata['BINQUANTITY'] = pd.cut(dataset['QUANTITYORDERED'], bins, labels=category, include_lowest=True)

newdata.head()
```

```python
newdata['BINQUANTITY'].value_counts()
```

*It turns out that the number of orders is dominated by the “Medium” class*

```python
quant_summary = newdata.groupby('BINQUANTITY').agg(row_cnt=('PRODUCTCODE','count'), min_quant=('QUANTITYORDERED','min'), max_quant=('QUANTITYORDERED','max'))
quant_summary['bin_interval'] = quant_summary['max_quant'] - quant_summary['min_quant']

quant_summary
```

*From the table above, it can be seen that the small bin has the smallest number of data rows (661), while the medium bin has the largest number of data rows (1796). It can also be seen that the small bin has the largest range of items (20.0), while the medium and large bins have the same range of items (19.0)*

### **PRICEEACH**
*The price value is divided into 3 bins, namely Small, Medium, and Large. So the results are as follows:*

```python
bins = np.linspace(min(dataset['PRICEEACH']), max(dataset['PRICEEACH']), 4)
category = ['Small', 'Medium', 'Large']

newdata['CATPRICE'] = pd.cut(dataset['PRICEEACH'], bins, labels=category, include_lowest=True)

newdata.head()
```

```python
newdata['CATPRICE'].value_counts()
```

*The results show that the cost of each product mostly falls into the Large class*

```python
price_summary = newdata.groupby('CATPRICE').agg(row_cnt=('PRODUCTCODE','count'), min_price=('PRICEEACH','min'), max_price=('PRICEEACH','max'))
price_summary['bin_interval'] = price_summary['max_price'] - price_summary['min_price']

price_summary
```

*From the table above, it can be seen that the small catprice has the smallest number of data rows (282), while the large catprice has the largest number of data rows (1916). It can also be seen that the small catprice has the largest price range (24.33), while the medium and large catprice have almost the same price range (24.23 and 24.37). In addition, we can see that the small catprice has a price between 26.88 and 51.21, the medium catprice has a price between 51.32 and 75.55, and the large catprice has a price between 75.63 and 100*

# **User Segmentation**
*To segment users or customers, in this process, grouping (groupby) is carried out on several variables. The bins used consist of three classes: Low, Medium, and High. In addition, there is a new column called SEGMENT, which displays user segmentation based on the number of transactions*

```python
newdata['TOTALSALES'] = newdata.groupby('CUSTOMERNAME')['SALES'].transform('sum')
```

```python
newdata['TRANSACTION_COUNT'] = newdata.groupby('CUSTOMERNAME')['SALES'].transform('count')
```

```python
newdata['AVG_ORDER_VALUE'] = newdata['TOTALSALES'] / newdata['TRANSACTION_COUNT']
```

```python
bins = [0, 100, 200, float('inf')]
labels = ['Low Value', 'Medium Value', 'High Value']

newdata['SEGMENT'] = pd.cut(newdata['TRANSACTION_COUNT'], bins=bins, labels=labels)
```

```python
usersegme = newdata[['CUSTOMERNAME', 'TOTALSALES', 'TRANSACTION_COUNT', 'AVG_ORDER_VALUE', 'SEGMENT']]

usersegme
```

*The table above shows the segmentation of each customer which is divided into three segments*

```python
newdata['SEGMENT'].value_counts()
```

```python
segment = newdata['SEGMENT'].value_counts()

plt.figure(figsize=(5,3))
sns.barplot(x=segment.index, y=segment.values, palette="dark:#5A9_r")

plt.xlabel('Segmentation')
plt.ylabel('Count')
plt.title('User Segmentation Distribution')
plt.xticks(rotation=45, ha='center')

plt.show()
```

*This bar chart shows that the low value segment dominates the number of users, while the medium and high value segments have much fewer users. This shows that there are challenges and opportunities to increase user value and loyalty with strategies that suit their segments*

```python
summary = pd.pivot_table(newdata, index='SEGMENT',
                         values=['CUSTOMERNAME', 'TOTALSALES', 'TRANSACTION_COUNT', 'AVG_ORDER_VALUE'],
                         aggfunc={'CUSTOMERNAME': pd.Series.nunique,
                                  'TOTALSALES': [np.mean, np.median],
                                  'TRANSACTION_COUNT': [np.mean, np.median],
                                  'AVG_ORDER_VALUE': [np.mean, np.median]})

summary['pct_unique'] = (summary['CUSTOMERNAME'] / summary['CUSTOMERNAME'].sum() * 100).round(1)

summary
```

*The table above shows that the high value segment has the highest value, while the low value segment has the lowest value. This shows that the high value segment is the most profitable, contributing, active, and loyal segment for the company, while the low value segment is the least profitable segment*
